```python
import pymysql
import pandas as pd
```

```python
conn = pymysql.connect(host="127.0.0.1", port=3306, db="apartment",
                user="root", passwd="1234")
```

```python
cursor = conn.cursor()
```

```python
metro_df = pd.read_csv("data-files/sub_2022.csv")
```

```python
# sql에 table 생성
sql = """
    create table metro
    (
        date char(8) not null,
        line_no varchar(10) not null,
        staion varchar(20) not null,
        get_on int not null,
        get_off int not null
    )
    """
cursor.execute(sql)
```

```python
# seoul_df를 list로 변환
metro_list = metro_df.values.tolist()
```

```python
cursor.execute("delete from metro")
conn.commit()
```
